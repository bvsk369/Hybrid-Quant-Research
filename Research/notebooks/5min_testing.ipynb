```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir("C:/Personal/Coding/Quant Project/Research")
df = pd.read_csv('Data/ICICIBANK_5minute_features.csv', index_col=0, parse_dates=True)
df.index.name = "timestamp"
from src.strategy_5min import backtest_mean_reversion_5min
```

```python
df=backtest_mean_reversion_5min(df)
```

```python
sharpe = (
    df['strategy_return'].mean() /
    df['strategy_return'].std()
) * np.sqrt(252 * 78)
```

```python
print("Mean:", df['strategy_return_net'].mean())
print("Std:", df['strategy_return_net'].std())
print("NaNs:", df['strategy_return_net'].isna().sum())
```

```python
ret = df['strategy_return_net'].dropna()

mean_ret = ret.mean()
std_ret = ret.std()

print("Mean return:", mean_ret)
print("Std return:", std_ret)

if std_ret == 0 or np.isnan(std_ret):
    print("Sharpe undefined (zero or NaN volatility)")
else:
    sharpe = (mean_ret / std_ret) * np.sqrt(252 * 78)
    print(f"Sharpe (net): {sharpe:.2f}")
```
