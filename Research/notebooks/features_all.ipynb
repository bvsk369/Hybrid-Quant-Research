```python
plt.figure(figsize=(8, 5))
plt.hist(df['zscore_20'], bins=100, density=True)

for z in [-3, -2, -1, 0, 1, 2, 3]:
    plt.axvline(z, linestyle='--')

plt.title("Z-Score Distribution with Thresholds")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.show()
```
